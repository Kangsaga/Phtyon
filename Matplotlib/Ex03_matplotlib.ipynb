### 선 그래프

```python
# plot 함수 - 일직선으로 되어있는 Line plot 그리기

y = [2, 4, 6]

plt.plot(y)   # 그래프 그리기
plt.show()   # 그래프 출력
```

```python
# x축 설정하여 그래프 그리기
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x,y)
plt.show()
```

```python
x = [2,4,6,8]
y = [10,15,18,20]

y2 = [10,20,16,13]

plt.plot(x,y)
plt.plot(x,y2, ls=':')
plt.show()
```

```python
plt.plot(x,y, marker='o')
plt.show()
```

```python
# line width -> lw
# 라인 두께 지정
plt.plot(x,y, linewidth=5)
plt.show()
```

```python
# marker color 색상변경
# markerfacecolor -> mfc
plt.plot(x,y,color='#086E7D' , marker = 'o', markerfacecolor = 'r')
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x,y, ls='--', color='g', marker='o')
plt.show()
```

```python
import numpy as np
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls='--', color='b',marker='o', mfc='r',
         markeredgecolor='g', ms=15, lw=5, mew =5)

plt.xlim(-2,8) # x 범위 지정 (최소, 최대)
plt.ylim(-1,11) # y 범위 지정 (최소, 최대)
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls='--', color='b',marker='o', mfc='r',
         markeredgecolor='g', ms=15, lw=5, mew =5)

plt.xticks([0,3,6])
plt.yticks([1,4,7])
plt.xlabel('x') # x축의 이름
plt.ylabel('y', rotation = 0) # y축의 이름
plt.grid()
plt.show()
```

```python
# 한 영역 안에 여러 그래프 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, color = 'r', label='x') # 그래프 이름
plt.plot(y, color = 'k', label='y')
plt.plot(z, color = 'm', label='z')
plt.legend(loc = 'best') # 범례 띄우기 loc = '영역'
plt.show()

# best
#   upper right
#   upper left
#   lower left
#   lower right
#   right
#   center left
#   center right
#   lower center
#   upper center
#   center
```

```python
plt.plot(x, color = 'r', label='엑스') # 그래프 이름
plt.plot(y, color = 'k', label='와이')
plt.plot(z, color = 'm', label='제트')
plt.xlabel('x축')
plt.ylabel('y축', rotation=0)
plt.title('x,y,z 그래프')
plt.legend(loc = 'best') # 범례 띄우기 loc = '영역'
plt.show()
```

#### 차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화

```python
# 1. 차대차 사건 중
# 2. (1)죽거나 다친 사람(사상자)이 많은 (2) 발생지 시도
# 3. 시각화
```

```python
data.head(2)
```

```python
acc = data[data['사고유형_대분류'] == '차대차']
```

```python
acc['발생지시도'].unique()
```

```python
acc['사상자수']
```

```python
# 그룹화 - groupby() 
acc_g = acc.loc[:,['사상자수','발생지시도']].groupby('발생지시도').sum()
acc_g
```

```python
# 시각화
x = acc_g.index
y = acc_g['사상자수']
```

```python
plt.figure(figsize = (10,5)) # 가로,세로 -> 그래프를 그리기 전에 사이즈를 정해놓아야 한다

plt.bar(x,y)
plt.title('차vs차 교통사고의 사상자 수')
plt.ylabel('사상자수', rotation = 0)
plt.xlabel('지역')

plt.show()
```
