```python
commentList = []
```

```python

for i in range(1,101):
    h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36' }
    driver.get('https://kin.naver.com/search/list.naver?query=%EC%86%90%ED%86%B1%20%EC%98%81%EC%96%91&page='+str(i))
    for i in range(10):
        
        title = driver.find_elements_by_css_selector("dt > a")[i]
        title.click()
        time.sleep(1)
        
        driver.switch_to.window(driver.window_handles[1])
        res = req.get(driver.current_url, headers = h)
        soup = bs(res.text, 'lxml')
        text = soup.select('div.question-content')[0].text
        text = text.replace('\n' , ' ')
        text = text.replace('\t' , ' ')
        commentList.append(text)
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(1)
```

```python
len(commentList)
```

```python
df = {'question' : commentList}
```

```python
data = pd.DataFrame(df)
```

```python
data.to_csv('손톱 영양.csv', encoding = 'utf-8-sig')
```
