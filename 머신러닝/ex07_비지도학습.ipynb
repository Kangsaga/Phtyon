## iris 데이터

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

### PCA
- 주성분 분석을 사용해서 4차원 데이터를 2차원으로 축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # 전체 데이터를 2차원으로 축소
```

```python
pca.fit(X)
```

```python
X_trans = pca.transform(X)
X_trans
```

```python
# 축소된 데이터 시각화
plt.figure(figsize=(10,10))
# 산점도
plt.scatter(X_trans[:,0], # 축소된 0번 컬럼
            X_trans[:,1], # 축소된 1번 컬럼
            c = y) 
plt.show()
```

```python
pca = PCA(n_components = 100, whiten= True)
```

```python
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
knn.fit(X_train_pca, y_train)
```

```python
knn.score(X_train_pca, y_train)
```

```python
knn.score(X_test_pca, y_test)
```

## 2차원 데이터 셋

```python
from sklearn.datasets import make_blobs # 2차원 데이터 셋 생성
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(random_state = 3)
```

```python
X.shape
```

```python
y
```

```python
# 군집 모델 생성
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
pred = kmeans.predict(X)
```

```python
pred
```

```python
# 중심 포인트의 좌표
```

```python
kmeans.cluster_centers_
```

```python
# 생성한 데이터가 나눈 그룹
for n_cluster in range(3):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
# KMeans 군집을 사용해서 나눈 그룹 시각화
center = kmeans.cluster_centers_
for n_cluster in range(3):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.scatter(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

```python
# KMeans
## 모든 방향의 중요도가 같다고 생각함
## 데이터가 특정한 모양을 가지고 있는 경우 그룹을 잘 만들지 못함
```

## 반달 데이터셋

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, random_state=0, noise=0.05)
```

```python
for n_cluster in range(2):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
pred = kmeans.predict(X)
```

```python
center = kmeans.cluster_centers_
for n_cluster in range(2):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

## 병합군집

```python
# 가까이 있는 것들을 그룹으로 삼는다
mglearn.plots.plot_agglomerative_algorithm()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg =  AgglomerativeClustering(n_clusters = 2)
pred = agg.fit_predict(X)
```

## DBSCAN
- 밀도 기반 클로스터링
- Density Based Spatial Clustering of Applications with noise

```python
# eps거리 안에 min_samples의 갯수를 가지면서 군집을 이른다
# 작은 세모나 원은 border point이며 큰 세모나 원은 core point
# 마지막으로 아무 색깔이 없는것은  noise point 이다
mglearn.plots.plot_dbscan()
```

```python
# DBSCAN은 스캐일링을 진해하는 것이 좋음
```

```python
from sklearn.preprocessing import StandardScaler
st_scaler = StandardScaler()
st_scaler.fit(X)
X_scaler = st_scaler.transform(X)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.5, min_samples=5)
pred = dbscan.fit_predict(X_scaler)
```

```python
pred
```

```python
# 데이터들이 모여있는 방향을 알 수 있음
for n_cluster in range(2):
  data = X[pred == n_cluster]
  plt.scatter(data[: , 0] , data[: , 1] , label = n_cluster)

plt.legend()
plt.show()
```
