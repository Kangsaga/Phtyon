```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys 
import time
import pandas as pd
import requests as req
from bs4 import BeautifulSoup as bs
import datetime
```

```python
driver = wb.Chrome()
```

```python
driver.get("https://www.youtube.com/results?search_query=%EC%82%B0%EB%B6%88")
```

```python
title = driver.find_elements_by_css_selector("#video-title")[0]
title.click()
```

```python
img = driver.find_elements_by_css_selector("#thumbnail img#img")[60]
img.click()
```

```python
soup = bs(driver.page_source,"lxml")
soup.select("#content-text")[5].text
```

```python
commentList = []

for i in range (110,120):
    if i % 5 == 0:
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        time.sleep(2)
        
    title = driver.find_elements_by_css_selector("#video-title")[i]
    title.click()
    time.sleep(2)
    
    body = driver.find_element_by_css_selector("body")
    try:
        for i in range(50):
            body.send_keys(Keys.END)
            time.sleep(1)
    except:
        print("댓글이 없습니다")
    soup = bs(driver.page_source,"lxml")
    comment = soup.select("#content-text")
    for cme in comment:
        commentList.append(cme.text)
    driver.back()
    time.sleep(1)
```

```python
with open('관련순 90-100.txt','w',encoding='UTF-8') as f:
    for cme in commentList:
        f.write(cme+'\n')
```
