## 이미지 불러오기

```python
# imread() : 외부 이미지 읽어욤
# cv2.IMREAD_COLOR : 이미지를 칼라 이미지로 설정
img = cv2.imread("./img/0.jpg",cv2.IMREAD_COLOR )
```

```python
import matplotlib.pyplot as plt

## 눈금은 삭제
plt.xticks([])
plt.yticks([])

plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

plt.xticks([])
plt.yticks([])

plt.imshow(img)
```
